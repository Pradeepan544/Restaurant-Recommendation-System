Analyzing different dishes across restaurants

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
data['dish_liked'] = data['dish_liked'].str.lower()
```

```python
data['dish_liked'] = data['dish_liked'].fillna('')
import re
def tokenize_dishes(text):
    # Split by commas outside quotes
    dishes = re.split(r',(?=(?:[^"]*"[^"]*")*[^"]*$)', text)
    # Remove leading/trailing whitespace and quotes
    dishes = [dish.strip('" ').strip() for dish in dishes]
    return dishes

data['dish_liked'] = data['dish_liked'].apply(tokenize_dishes)
print(data['dish_liked'])
```

```python
all_dishes = []
for dishes in data['dish_liked']:
    all_dishes.extend(dishes)
dish_counts = pd.Series(all_dishes).value_counts()
dish_counts = dish_counts[dish_counts.index != '']
dish_counts = dish_counts[dish_counts.index.str.isnumeric() == False]

# Visualize top 10 popular dishes
dish_counts.head(50).plot(kind='bar', figsize=(15, 6))
```

**Location Analysis**

```python
data['cuisines'] = data['cuisines'].str.lower().str.strip()
data['cuisines'] = data['cuisines'].str.split(', ')
data = data.explode('cuisines')
data['location']=data['location'].str.title()
popular_cuisines = data['cuisines'].value_counts().head(20).index.tolist()
print(popular_cuisines)
```

```python
top_locations_by_cuisine = {}
for cuisine in popular_cuisines:
    cuisine_df = data[data['cuisines'] == cuisine]
    top_locations = cuisine_df['location'].value_counts().head(5).index.tolist()
    top_locations_by_cuisine[cuisine] = top_locations

# Print the results
for cuisine, locations in top_locations_by_cuisine.items():
    print(f"Top Locations for {cuisine.title()}:")
    for location in locations:
        print(f"- {location}")
```

**Cuisine and Restaurant Type Pairing**

```python
cuisine_counts = data.groupby(['rest_type', 'cuisines']).size().reset_index(name='count')
cuisine_counts = cuisine_counts.sort_values(['rest_type', 'count'], ascending=[True, False])
print(cuisine_counts)
```

```python
plt.figure(figsize=(14, 70))
top_cuisines = cuisine_counts[cuisine_counts['count'] > 1]  # Filter to show only popular cuisines
sns.barplot(data=top_cuisines, x='count', y='cuisines', hue='rest_type')
plt.title('Cuisine Popularity by Restaurant Type')
plt.xlabel('Count of Cuisines')
plt.ylabel('Cuisine')
plt.show()
```
