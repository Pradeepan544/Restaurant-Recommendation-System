```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('zomato_small_40%.csv', encoding='cp1252')
df.head()
```

```python
df = df.iloc[:, :17]
```

```python
df=df.dropna(how='all')
```

```python
df
```

```python
df_cleaned = df[
    df['url'].str.startswith('https://') |
    df['url'].str.startswith('http://')
]
```

```python
df_cleaned.shape
```

```python
df_cleaned.head()
```

```python
df_cleaned.describe()
```

```python

missing = pd.DataFrame((df_cleaned.isnull().sum())*100/df_cleaned.shape[0]).reset_index()
plt.figure(figsize=(16,5))
ax = sns.pointplot(x='index',y=0,data=missing)
plt.xticks(rotation =90,fontsize =7)
plt.title("Percentage of Missing values")
plt.ylabel("PERCENTAGE")
plt.show()
```

Handling missing values

```python
#distribution of Rate
plt.figure(figsize=(12, 6))
sns.histplot(df_cleaned['rate'].dropna(), kde=True)
plt.title('Distribution of Rate')
plt.xlabel('Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
df_cleaned['rate'] = df_cleaned['rate'].str.replace('/5', '').str.replace(' ', '')
```

```python
df_cleaned['rate'] = pd.to_numeric(df_cleaned['rate'], errors='coerce')
```

```python
median_rate = df_cleaned['rate'].median()

# Step 4: Fill missing values in the original 'rate' column with the median as a numeric value
df_cleaned['rate'] = df_cleaned['rate'].fillna(median_rate)
```

```python
#handling menu_item
df_cleaned['menu_item'].fillna('Not Available', inplace=True)
```

```python
#handling listed_in(type)
df_cleaned['listed_in(type)'].fillna('Not Specified', inplace=True)
```

```python
#handling listed_in(city)
df_cleaned['listed_in(city)'].fillna('Not Specified', inplace=True)
```

```python
missing = pd.DataFrame((df_cleaned.isnull().sum())*100/df_cleaned.shape[0]).reset_index()
plt.figure(figsize=(16,5))
ax = sns.pointplot(x='index',y=0,data=missing)
plt.xticks(rotation =90,fontsize =7)
plt.title("Percentage of Missing values")
plt.ylabel("PERCENTAGE")
plt.show()
```

```python
df_cleaned.to_csv('zomato_cleaned_40%.csv', index=False)
```
